```python
# Naive Solution

example_string1 = "the quick brown fox jumps over a lazy dog"

all_alphabets = True

for i in range(97, 123):
    print(chr(i))
    found = False
    for char in example_string1:
        if char == chr(i):
            found = True
            print(f'alphabet {chr(i)} is present')
    if found == False:
        all_alphabets = False
        print(f'alphabet {chr(i)} is NOT present')
        break

print(f"all_alphabets {all_alphabets}")
```

```python
alphabet_sheet = []
for i in range(97, 123):
    alphabet_sheet.append(None)
```

```python
example_string1 = "the quick brown fox jumps over a lazy dog"

for stu in example_string1:
    index_position = ord(stu)-97
    if index_position >= 0 and index_position <= 25:
        alphabet_sheet[index_position] = 1

print(alphabet_sheet)
```

```python
all_alphabets = True
for stu in alphabet_sheet:
    if stu is None:
        all_alphabets = False
print(all_alphabets)
```
