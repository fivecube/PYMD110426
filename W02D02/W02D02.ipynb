```python
data_list = []
data_dict = {}

for i in range(1, 100000000):
    data_list.append(i)
    data_dict[i] = True
```

```python
key_to_search = 90809008
```

```python
import time

start = time.perf_counter()

for no in data_list:
    if no == key_to_search:
        print('found')
        break

end = time.perf_counter()

duration = end - start

print(f"{duration:.4f} seconds")
```

```python
import time

start = time.perf_counter()

result = data_dict.get(key_to_search)

print('found', result)

end = time.perf_counter()

duration = end - start

print(f"{duration:.4f} seconds")
```

```python
students = [
            {"name": "Vinay", "last_name": "Khanna", "percentage": 78},
            {"last_name": "Shukla", "percentage": 92, "name": "Praveen"}
           ]
```

```python
for stu in students:
    print(stu)
```

```python
for stu in students:
    print(f"Student Name -> {stu.get('name') + ' ' + stu.get('last_name')}. % = {stu.get('percentage')}")
```

```python
def calculate_percentage(list_of_marks):
    number_of_subjects = len(list_of_marks)
    total_marks = sum(list_of_marks)
    percentage = total_marks/number_of_subjects
    return percentage
```

```python
students = [
            {"name": "Vinay", "last_name": "Khanna", "percentage": "nothing",
             "marks": [45, 60, 39]
            },
    
            {"last_name": "Shukla", "percentage": "nothing", "name": "Praveen",
            "marks": [49, 67, 53]
            }
           ]
```

```python
for stu in students:
    per = calculate_percentage(stu.get('marks'))
    stu['percentage'] = per
    print(f"Student Name -> {stu.get('name') + ' ' + stu.get('last_name')}. % = {stu.get('percentage')}")
```

```python
students
```

```python
students = [
            {"name": "Vinay", "last_name": "Khanna", "percentage": "nothing",
             "marks": [
                        {'name': 'maths', 'value': 45}, 
                        {'name': 'english', 'value': 60}, 
                        {'name': 'french', 'value': 39}
                      ]
            },
    
            {"last_name": "Shukla", "percentage": "nothing", "name": "Praveen",
            "marks": [
                        {'name': 'sankskrit', 'value': 49},
                        {'name': 'hindi', 'value': 67},
                        {'name': 'kannada', 'value': 53},
                     ]
            }
           ]
```

```python
for stu in students:
    highest_marks = float('-inf')
    marks_list = []
    hi_subject = None
    for ma_dict in stu.get('marks'):
        marks_list.append(ma_dict.get('value'))
        if ma_dict.get('value') > highest_marks:
            highest_marks = ma_dict.get('value')
            hi_subject = ma_dict.get('name')

    per = calculate_percentage(marks_list)
    stu['percentage'] = per
    print(f"Student Name -> {stu.get('name') + ' ' + stu.get('last_name')}. % = {stu.get('percentage')}. \
    Highest Marks -> {highest_marks} and in subject = {hi_subject}")
```

```python
# Classes and objects
# Class methods and static methods


class Vehicle:
    def __init__(self, name, model):
        self.name = name
        self.model = model

    def get_name(self):
        return self.name

    def set_name(self, name):
        self.name = name
```

```python
obj1 = Vehicle('My Car 1', 'Tata Nexon')
```
