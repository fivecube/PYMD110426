```python
# What is the self ?

1) It is basically just a name which everybody agreed on, but if you wish you can change.
2) What matters is it's position, It's always the first argument in object method/function.
3) Whenever you are calling an object method, self is automatically passed as the reference to the object.
4) Inside the object method, you will be able to access properties/attributes of object because you have
reference to the object.
5) self/first argument is implicitly passed (which means we don't see it getting passed, It's auto done by Python)
```

```python
# class methods:- These are methods/functions that you use to access class attributes and modify.

# decorator classmethod is used for creating and distinguishing class methods from object methods.

# cls keyword:- The first argument in class method is reserved for class reference.

# cls is also implicit passed, and you can call class methods from objects, class as well. Because
# Python implicitly passes the class reference auto matically for us.
```

```python
class Vehicle:
    generation = 'BE6'  # class attribute

    def __init__(self, name, model):
        self.name = name  # self.name is acessing an object attribute name
        self.model = model  # self.model is accessing an object attribute model

    def get_name(self):
        return self.name

    def set_name(self, name):
        self.name = name

    @classmethod
    def get_generation(cls):
        return cls.generation

    @classmethod
    def set_generation(cls, generation):
        cls.generation = generation
```

```python
# Python does not enforce in a very hard manner for private, protected variables.
# But still have some ways to do similar things.
    
# If you use _ before any attr/var it will be treated as protected
# If you use __ before any attr/var it will be treated as private

# But there is a way to bypass this and access private attrs even outside the class and that is why
# we say python does not enforce this.
```

```python
class Vehicle:
    generation = 'BE6'  # class attribute

    def __init__(self, name, model):
        self.__name = name  # self.__name is private attr
        self.__model = model  # self.__model is private attr

    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

    @classmethod
    def get_generation(cls):
        return cls.generation

    @classmethod
    def set_generation(cls, generation):
        cls.generation = generation
```

```python
# staticmethod contains code which is not modifying any object attribute or class attribute.
# also it does not access object attribute directly with self and class attribute cls reference.
```

```python
class Vehicle:
    generation = 'BE6'  # class attribute

    def __init__(self, name, model):
        self.__name = name  # self.__name is private attr
        self.__model = model  # self.__model is private attr

    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

    @classmethod
    def get_generation(cls):
        return cls.generation

    @classmethod
    def set_generation(cls, generation):
        cls.generation = generation

    @staticmethod
    def calculate_mileage(distance, fuel):
        if fuel <= 0:
            return "Fuel must be greater than zero."
        mileage = distance / fuel
        return round(mileage, 2)
```

```python
class Coordinate:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"{self.x}, {self.y}"

    def __add__(self, obj2):
        new_x = self.x + obj2.x
        new_y = self.y + obj2.y
        new_object = Coordinate(new_x, new_y)
        return new_object
```

```python
class Vehicle:
    generation = 'BE6'  # class attribute

    def __init__(self, name, model):
        self.__name = name  # self.__name is private attr
        self.__model = model  # self.__model is private attr
        self.__speed = 0

    def get_name(self):
        return self.__name

    def set_name(self, name):
        self.__name = name

    @classmethod
    def get_generation(cls):
        return cls.generation

    @classmethod
    def set_generation(cls, generation):
        cls.generation = generation

    @staticmethod
    def calculate_mileage(distance, fuel):
        if fuel <= 0:
            return "Fuel must be greater than zero."
        mileage = distance / fuel
        return round(mileage, 2)
```

```python
class Car(Vehicle):
    def __init__(self, no_of_seatbelt, steering_mode, name, model):
        self.no_of_seatbelt = no_of_seatbelt
        self.steering_mode = steering_mode
        super().__init__(name, model)

    def accelerate(self, acc_pwr):
        self.__speed += acc_pwr*10

    def brake(self, brake_pwr):
        self.__speed -= brake_pwr*20
```

```python
my_office_car = Car(4, 'Auto', 'My Office Car', 'Tata Nexon')
```
