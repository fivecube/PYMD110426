```python
# Modes of file operations
"r" -> read
File opens in reading mode, you can not write to a file.
"w" -> writing
File opens in writing mode, you can not read from a file.
If the file exists, it will be overwritten.
Also If the file does not exist, new file will be created with the name.


"a" -> append (write at the end)
"r+" -> read + write
```

```python
try:
    num = int(input('Enter a number'))
    print(10/num)
except ZeroDivisionError:
    print('Can not divide by 0')
except ValueError:
    print('Please input only number')

print("let's do next activity")
```

```python
list_of_fruits = ['apple', 'mango']
list_of_fruits[3]
```

```python
try:
    num = int(input('Enter a number'))
    list_of_fruits = ['apple', 'mango']
    print(10/num)
    print(list_of_fruits[num])
except ZeroDivisionError:
    print('Can not divide by 0')
except ValueError:
    print('Please input only number')
except Exception as exc:
    print(type(exc))
    print(f'There is an issue, please try later! {exc}')

print("let's do next activity")
```

```python
try:
    num = int(input('Enter a number'))
    list_of_fruits = ['apple', 'mango']
    print(10/num)
    print(list_of_fruits[num])
except ZeroDivisionError:
    print('Can not divide by 0')
except ValueError:
    print('Please input only number')
except Exception as exc:
    print(type(exc))
    print(f'There is an issue, please try later! {exc}')
finally:
    print('We are done with exception handling')
print("let's do next activity")
```
