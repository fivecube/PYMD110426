```python
# Function definition and calling
# function:- grouping some lines of code to be called again and again anywhere in your code cycle.

# Keywords in python:-
# reserved words and you can not use them for variable names
# for, if, else, elif, def, len, sum, continue, break, etc
```

```python
# Built in function len
list_of_marks = [50, 30, 77, 93]
number_of_subjects = len(list_of_marks)
print(number_of_subjects)
```

```python
# Build in function sum
list_of_marks = [50, 30, 77, 93]
total_marks = sum(list_of_marks)
print(total_marks)
```

```python
def calculate_percentage(list_of_marks):
    number_of_subjects = len(list_of_marks)
    total_marks = sum(list_of_marks)
    percentage = total_marks/number_of_subjects
    return percentage
```

```python
# Once a function is defined, you can call it anywhere.
# For calling you are either in the same python file or in a different python file.
# if you are on a diff file, you import it from the other file where the fun is present.
```

```python
calculate_percentage([45, 67, 89])
```

```python
calculate_percentage([78, 23, 92, 99, 88])
```
